```python
def get_mean(nums):
    return sum(nums) / len(nums)

def get_median(nums):
    nums = sorted(nums)
    n = len(nums)
    mid = n // 2

    if n % 2 == 1:
        return nums[mid]
    else:
        return (nums[mid - 1] + nums[mid]) / 2

def get_mode(nums):
    freq = {}
    for x in nums:
        freq[x] = freq.get(x, 0) + 1

    max_count = max(freq.values())
    modes = [num for num, count in freq.items() if count == max_count]

    if len(modes) == 1:
        return modes[0]
    else:
        return "No unique mode (multiple values share the same highest frequency)"

count = int(input("How many participants? "))

values = []

for i in range(count):
    num = float(input(f"Enter value for participant {i+1}: "))
    values.append(num)

print("\nWhat do you want to calculate?")
print("[1] Mean")
print("[2] Median")
print("[3] Mode")

choice = int(input("Choose 1/2/3: "))

if choice == 1:
    print("\nMean:", get_mean(values))
elif choice == 2:
    print("\nMedian:", get_median(values))
elif choice == 3:
    print("\nMode:", get_mode(values))
else:
    print("Invalid choice.")
```
